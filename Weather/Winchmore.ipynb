```python
import datetime
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
RawData = pd.read_csv('Winchmore.met',skiprows=11,usecols=[1,4,5,7,8,10,19],
                      names=['DATE','WIND','RAIN','TMAX','TMIN','VPRS','SRAD'],
                     na_values = ['-'])
RawData.ix[:,'DATE'] =  pd.to_datetime(RawData.ix[:,'DATE'])
RawData.set_index('DATE',inplace=True,drop=False)
RawData.ix[:,'WIND'] = RawData.ix[:,'WIND'] * 1000 / (24 * 60 * 60)
idxSer = range(RawData.index.size)
RawData.ix[:,'Year'] = [RawData.ix[x,'DATE'].year for x in idxSer]
RawData.ix[:,'Day'] = [RawData.ix[x,'DATE'].dayofyear for x in idxSer]
```

```python
Fig = plt.figure(figsize = (18,24))
Position = 1
for x in ['TMIN', 'TMAX', 'RAIN', u'SRAD', u'WIND', u'VPRS']:
    Fig.add_subplot(3,2,Position)
    plt.plot(RawData.ix[:,'Day'],RawData.ix[:,x],'o')
    plt.ylabel(x,fontsize = 16)
    Position +=1
```
